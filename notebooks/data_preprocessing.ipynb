```python
import pandas as pd
import string
import re

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data/fdic_paragraphs_100_token.csv")
df.shape
```

```python
def preprocess_text(text: str):
  text = re.sub("\([a-zA-z0-9]{1,2}\)", "", text)
  text = text.encode("ascii", errors="ignore").decode()
  text = re.sub("[^a-zA-Z\s](?!(?<=[a-zA-Z].)[a-zA-Z])", "", text)
  text = re.sub("\s{2,}|\t", " ", text)
  return text.strip()
  
```

```python
for i, t in enumerate(df["TEXT"].tolist()):
  try:
    preprocess_text(t)
  except:
    print(i)

```

```python
df["preprocessed_text"] = df["TEXT"].apply(preprocess_text)
```
