```python
import pandas as pd
from sentence_transformers import SentenceTransformer

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data/fdic_paragraphs_improved.csv")
```

```python
model_roberta = SentenceTransformer('/content/drive/MyDrive/Colab Notebooks/models/finetuned_roberta')
model_mpnet = SentenceTransformer('/content/drive/MyDrive/Colab Notebooks/models/finetuned_mpnet')
```

```python
output_vec_improved_passage = model_roberta.encode(df["TEXT"], show_progress_bar = True)
```

```python
output_vectors_roberta = model_roberta.encode(df["preprocessed_text"], show_progress_bar = True)
output_vectors_mpnet = model_mpnet.encode(df["preprocessed_text"], show_progress_bar = True)
```

```python
pd.DataFrame(output_vectors_roberta).to_csv('/content/drive/MyDrive/Colab Notebooks/data/stored_vectors_roberta.csv', index = False)
pd.DataFrame(output_vectors_mpnet).to_csv('/content/drive/MyDrive/Colab Notebooks/data/stored_vectors_mpnet.csv', index = False)
```

```python
pd.DataFrame(output_vec_improved_passage).to_csv('/content/drive/MyDrive/Colab Notebooks/data/stored_vectors_roberta_improved.csv', index = False)
```
