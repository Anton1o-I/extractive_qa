```python
file_information = []
root = file.getroot()
document = root.iter("SECTION")

max_len = 150

#loop through the section identifier in the xml
for p in document:
  sectno = p.find("SECTNO").text
  subject = "" if p.find("SUBJECT") is None else p.find("SUBJECT").text
  section_text = []
  #loop through the paragraph identifiers in xml
  for i, para in enumerate(p.iter("P")):
    text = " ".join([t for t in para.itertext()])
    section_text.append(text)
  #loop through paragraph chunks, if chunk doesn't end in '.' and is followed by () combine sentences 
  passage = ''
  counter = 1
  for i, sen in enumerate(section_text):
    current_sen = remove_whitespace(section_text[i])
    passage += " " + current_sen
    if current_sen[-1] != ".":
      continue
    else:
      out = {
          "SECTNO": sectno,
          "SUBJECT": subject,
          "PARAGRAPH": counter,
          "TEXT": passage.strip()
      }
      counter+=1
      passage = ''
      file_information.append(out)
```

```python
clean_file2 = pd.DataFrame(file_information)
clean_file2.to_csv("/content/drive/MyDrive/Colab Notebooks/data/fdic_paragraphs_improved.csv", index = False)
```
